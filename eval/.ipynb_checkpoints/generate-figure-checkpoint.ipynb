```python
%config IPCompleter.greedy=True
%matplotlib inline
import csv
from matplotlib2tikz import save as tikz_save
```

```python
'''Read CSV'''
raw_data = {}

with open('results.csv') as data_file:
    reader = csv.DictReader(data_file)
    for row in reader:
        bench_name = row['bench'].strip()
        config_name = row[' config'].strip()
        overhead_value = row[' overhead'].strip()
        if bench_name not in raw_data:
            raw_data[bench_name] = {}
        raw_data[bench_name][config_name] = overhead_value

## Combine with Parrot paper result
with open('parrot-paper-results.csv') as data_file1:
    reader = csv.DictReader(data_file1)
    for row in reader:
        bench_name = row['bench']
        if bench_name in raw_data:
            raw_data[bench_name]['Parrot paper result']=row['Parrot paper result']
```

```python
## list to filter benchmark with pcs
pcs_list = ["splash2x cholesky",
"splash2x fmm",
"splash2x raytrace",
"pfscan",
"npb ua-l",
"parsec fluidanimate"]
```

```python
'''Choose no-pcs-hint for all benchmarks'''

for bench in raw_data:
    if 'xtern-no-pcs-hint' not in raw_data[bench]:
        if bench in pcs_list or 'xtern-hinted' not in raw_data[bench]:
            raw_data[bench]['xtern-no-pcs-hint']=raw_data[bench]['xtern-no-hint']
        else:
            raw_data[bench]['xtern-no-pcs-hint']=raw_data[bench]['xtern-hinted']
```

```python
def unformat_percentage(val):
    if (val == 'N/A'):
        return 0
    assert val.endswith('%'), "Unexpected val: %s"%val
    return (float(val[0:-1])+100)/100

def get_relative_timings(benchmarks, column, filter_list=None):
    vals = []
    for item in benchmarks:
        if (item == "GAP" or (filter_list and item not in filter_list)):
            vals.append(0)
        elif item in raw_data:
            vals.append(unformat_percentage(raw_data[item][column]))
        else:
            vals.append(0)
    return vals
```

```python
def addBreakClips(ax1,ax2):
    """ Code to add diagonal slashes to truncated y-axes.
    copied from http://matplotlib.org/examples/pylab_examples/broken_axis.html"""
    
    d = .004 # how big to make the diagonal lines in axes coordinates
    # arguments to pass plot, just so we don't keep repeating them
    kwargs = dict(transform=ax1.transAxes, color="black", clip_on=False)
    ax1.plot((-d,+d),(-d,+d), **kwargs)      # top-left diagonal
    ax1.plot((1-d,1+d),(-d,+d), **kwargs)    # top-right diagonal

    kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
    ax2.plot((-d,+d),(1-d,1+d), **kwargs)   # bottom-left diagonal
    ax2.plot((1-d,1+d),(1-d,1+d), **kwargs) # bottom-right diagonal
```

```python
def plot_figure(benchmarks, val1, val2, val3, val4_stacked, 
                val1_caption="", val2_caption="", val3_caption="", val4_caption=""):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib import gridspec
    import matplotlib.ticker as ticker
    
    N = len(benchmarks)
    width = 0.17
    ind = np.array([i*width*4 for i in range(N)])
    plt.rcParams['text.usetex'] = True
    plt.rcParams["font.family"] = "serif"
    plt.rcParams["font.serif"] = ["Linux Libertine"]
    plt.rcParams['font.size'] = 12
    plt.rcParams['text.latex.preamble'] = r'\usepackage{libertine}'
    
    fig = plt.figure(figsize=(16,3))
    gs = gridspec.GridSpec(2, 1, height_ratios=[1, 2])
    ax_upper = fig.add_subplot(gs[0])
    ax_lower = fig.add_subplot(gs[1], sharex=ax_upper)
#   fig,(ax_upper,ax_lower) = plt.subplots(2, 1, figsize=(14,4), sharex=True)
    #squeeze plots closer
    plt.subplots_adjust(hspace=0.1)
    addBreakClips(ax_upper,ax_lower)
       
    upper_lower_bound = 20
    
    ax_upper.set_ylim(upper_lower_bound,200)
    ax_upper.set_xlim(0-width*2, N*width*4)
    ax_lower.set_ylim(0,10)
    ax_upper.set_yscale('log')
    
    #remove the bottom border from the top plot and the upper border from the bottom plot
    ax_upper.spines['bottom'].set_visible(False)
    ax_upper.spines['top'].set_visible(False)
    ax_lower.spines['top'].set_visible(False)
    
    
    ax_lower.tick_params(bottom=False, labelright=True) # hides the labels from the top x-axis
    ax_upper.tick_params(bottom=False, labelright=True) # hides the labels from the top x-axis
    ax_upper.yaxis.set_minor_formatter(ticker.NullFormatter())
    ax_upper.yaxis.set_major_formatter(ticker.ScalarFormatter())
    ax_upper.set_yticks([20,100,200])
    
    # Add a line at y=1
    ax_lower.set_yticks([1], minor=True)
    ax_lower.yaxis.grid(True, which='minor', zorder=0)
    
    rects_val1 = ax_lower.bar(ind, val1, width, edgecolor='black', color='none', hatch="///", zorder=3)
    rects_val2 = ax_lower.bar(ind+width, val2, width, edgecolor='black', color='none', hatch="..", zorder=3)
    rects_val3 = ax_lower.bar(ind+width*2, val3, width, edgecolor='black', color='none', hatch="", zorder=3)
    rects_val4 = ax_lower.bar(ind+width, val4_stacked, width, edgecolor='black', color='black', zorder=3)
    
    rects_val1 = ax_upper.bar(ind, val1, width, edgecolor='black', color='none', hatch="///", zorder=3)
    rects_val2 = ax_upper.bar(ind+width, val2, width, edgecolor='black', color='none', hatch="..", zorder=3)
    rects_val3 = ax_upper.bar(ind+width*2, val3, width, edgecolor='black', color='none', hatch="", zorder=3)
    rects_val4 = ax_upper.bar(ind+width, val4_stacked, width, edgecolor='black', color='black', zorder=3)
    
    ax_lower.set_xticks(ind+width)
    ax_lower.set_xticklabels(benchmarks, rotation=-90)
    l = ax_upper.legend((rects_val1[0], rects_val2[0], rects_val4[0], rects_val3[0] ), 
             (val1_caption, val2_caption, val4_caption, val3_caption ), frameon=False, ncol=2,
              loc="lower left", bbox_to_anchor=(0,0.3))
    l.set_zorder(20)
    
    plt.setp(ax_upper.get_xticklabels(), visible=False)
    
    def autolabel(rects):
        for rect in rects:
            h = rect.get_height()
            if (h>upper_lower_bound):
                ax_upper.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                    ha='center', va='bottom', size='7')
    
    #autolabel(rects_val1)
    #autolabel(rects_val2)
    #autolabel(rects_val3)
    
    return plt
```

```python
def plot_figure2(benchmarks, val1, val2, val3, val4_stacked, 
                     val1_caption="", val2_caption="", val3_caption="", val4_caption=""):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib import gridspec
    import matplotlib.ticker as ticker
    
    N = len(benchmarks)
    width = 0.17
    ind = np.array([i*width*4 for i in range(N)])
    plt.rcParams['text.usetex'] = True
    plt.rcParams["font.family"] = "serif"
    plt.rcParams["font.serif"] = ["Linux Libertine"]
    plt.rcParams['font.size'] = 12
    plt.rcParams['text.latex.preamble'] = r'\usepackage{libertine}'
    
    fig = plt.figure(figsize=(16,1.5))
    gs = gridspec.GridSpec(2, 1, height_ratios=[1, 2])
    ax_lower = fig.add_subplot(gs[1])
#   fig,(ax_upper,ax_lower) = plt.subplots(2, 1, figsize=(14,4), sharex=True)
    
    ax_lower.set_ylim(0,5)
    ax_lower.set_xlim(0-width*2, N*width*4)
    
    # Add a line at y=1
    ax_lower.set_yticks([1], minor=True)
    ax_lower.yaxis.grid(True, which='minor', zorder=0)
    
    #remove the bottom border from the top plot and the upper border from the bottom plot
    ax_lower.spines['top'].set_visible(False)
    
    ax_lower.tick_params(bottom=False, labelright=True) # hides the labels from the top x-axis
    
    rects_val1 = ax_lower.bar(ind, val1, width, edgecolor='black', color='none', hatch="///", zorder=3)
    rects_val2 = ax_lower.bar(ind+width, val2, width, edgecolor='black', color='none', hatch="..", zorder=3)
    rects_val3 = ax_lower.bar(ind+width*2, val3, width, edgecolor='black', color='none', hatch="", zorder=3)
    rects_val4 = ax_lower.bar(ind+width, val4_stacked, width, edgecolor='black', color='black', zorder=3)
    
    ax_lower.set_xticks(ind+width)
    ax_lower.set_xticklabels(benchmarks, rotation=-90)
#     ax_lower.legend((rects_val1[0], rects_val2[0], rects_val3[0]), 
#              (val1_caption, val2_caption, val3_caption))
    
    
    def autolabel(rects):
        for rect in rects:
            h = rect.get_height()
            if (h>upper_lower_bound):
                ax_upper.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                    ha='center', va='bottom', size='7')
    
    #autolabel(rects_val1)
    #autolabel(rects_val2)
    #autolabel(rects_val3)
    
    return plt
```
